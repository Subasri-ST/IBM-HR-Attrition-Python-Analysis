```python
import pandas as pd        
import numpy as np         
import matplotlib.pyplot as plt  
%matplotlib inline         
print("All libraries loaded successfully!")
```

```python
# Loading the IBM HR dataset into Python
df = pd.read_csv('C:/Users/LENOVO/Downloads/IBM-dataset-raw.csv')

# Check it loaded correctly
print("Dataset loaded successfully!")
print("Rows and Columns:", df.shape)
```

```python
# See the first 5 rows of data
print("First 5 rows:")
df.head()
```

```python
# See all column names
print("All column names:")
print(df.columns.tolist())

# See how many rows and columns
print("\nShape of data:")
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

```python
# Check data type of each column
print("Data types of each column:")
print(df.dtypes)
```

```python
# Check if any data is missing
print("Missing values in each column:")
print(df.isnull().sum())

# Total missing values in entire dataset
print("\nTotal missing values:", df.isnull().sum().sum())
```

```python
# Get basic statistics of all numeric columns
print("Basic statistics:")
df.describe()
```

```python
# First check columns with same value for everyone
print("EmployeeCount unique values:", df['EmployeeCount'].unique())
print("StandardHours unique values:", df['StandardHours'].unique())
print("Over18 unique values:", df['Over18'].unique())
```

```python
# Remove useless columns
df = df.drop(columns=['EmployeeCount', 'StandardHours', 'Over18'])

# Confirm they are removed
print("Columns after cleaning:")
print("New shape:", df.shape)
```

```python
# Rename columns to clearer names
df = df.rename(columns={
    'NumCompaniesWorked' : 'PreviousCompanies',
    'TotalWorkingYears'  : 'TotalExperience',
    'DistanceFromHome'   : 'CommuteDistance'
})

# Confirm renaming worked
print("Columns renamed successfully!")
print(df.columns.tolist())
```

```python
# Check for duplicate rows
print("Duplicate rows:", df.duplicated().sum())

# Check unique values in key columns
print("\nUnique Departments:", df['Department'].unique())
print("Unique Job Roles:", df['JobRole'].unique())
print("Unique Attrition values:", df['Attrition'].unique())
```

```python
# Convert Yes/No text to 1/0 numbers
# This helps in calculations and analysis

df['AttritionFlag'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# Confirm it worked
print("Attrition encoding:")
print(df['AttritionFlag'].value_counts())
print("\nAttrition Rate:", df['AttritionFlag'].mean().round(4) * 100, "%")
```

```python
# Save cleaned dataset to your computer
df.to_csv('C:/Users/LENOVO/Downloads/IBM_HR_Cleaned.csv', index=False)

print("✅ Cleaned data saved!")
print("Original columns: 35")
print("Cleaned columns:", df.shape[1])
print("Rows:", df.shape[0])
print("\nData Cleaning COMPLETE!")
```

```python
# Convert salary column to numpy array
salary = np.array(df['MonthlyIncome'])

# Basic statistics
print("=== Monthly Salary Statistics ===")
print("Average salary  :", np.mean(salary).round(2))
print("Middle salary   :", np.median(salary).round(2))
print("Lowest salary   :", np.min(salary))
print("Highest salary  :", np.max(salary))
print("Std deviation   :", np.std(salary).round(2))
```

```python
# Salary percentiles - who earns what
print("=== Salary Percentiles ===")
print("Bottom 25% earn below:", np.percentile(salary, 25))
print("Middle 50% earn below:", np.percentile(salary, 50))
print("Top 25% earn above   :", np.percentile(salary, 75))
print("Top 10% earn above   :", np.percentile(salary, 90))
```

```python
# Create salary category using np.where
df['SalaryCategory'] = np.where(
    df['MonthlyIncome'] < 4919,
    'Low Salary',
    'High Salary'
)
# See the result
print("=== Salary Categories ===")
print(df['SalaryCategory'].value_counts())
```

```python
# Check relationship between two columns
print("=== Correlation Analysis ===")

# Age vs Salary
corr1 = np.corrcoef(df['Age'], df['MonthlyIncome'])[0,1]
print("Age & Salary correlation      :", corr1.round(3))

# Experience vs Salary
corr2 = np.corrcoef(df['TotalExperience'], df['MonthlyIncome'])[0,1]
print("Experience & Salary correlation:", corr2.round(3))

# Commute vs Attrition
corr3 = np.corrcoef(df['CommuteDistance'], df['AttritionFlag'])[0,1]
print("Commute & Attrition correlation:", corr3.round(3))

print("\nScale: -1 = opposite | 0 = no relation | 1 = perfect match")
```

```python
# Attrition count by Department
print("=== Attrition by Department ===")
dept = df.groupby('Department')['AttritionFlag'].sum()
print(dept)

print("\n=== Attrition RATE by Department ===")
dept_rate = df.groupby('Department')['AttritionFlag'].mean() * 100
print(dept_rate.round(2))
```

```python
# Create age groups
df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=[18, 25, 35, 45, 60],
    labels=['18-25', '26-35', '36-45', '46-60']
)

# Attrition by age group
print("=== Attrition by Age Group ===")
age_attrition = df.groupby('AgeGroup',observed=False)['AttritionFlag'].mean() * 100
print(age_attrition.round(2))
```

```python
# Attrition by salary category
print("=== Attrition by Salary Category ===")
salary_attrition = df.groupby('SalaryCategory')['AttritionFlag'].mean() * 100
print(salary_attrition.round(2))

# Count of employees in each category
print("\n=== Employee Count by Salary Category ===")
print(df.groupby('SalaryCategory')['AttritionFlag'].count())
```

```python
# Attrition by overtime
print("=== Attrition by Overtime ===")
overtime = df.groupby('OverTime')['AttritionFlag'].mean() * 100
print(overtime.round(2))

print("\n=== Employee Count by Overtime ===")
print(df.groupby('OverTime')['AttritionFlag'].count())
```

```python
# Attrition rate by job role
print("=== Attrition Rate by Job Role ===")
jobrole = df.groupby('JobRole')['AttritionFlag'].mean() * 100
jobrole = jobrole.sort_values(ascending=False)
print(jobrole.round(2))
```

```python
# Pie chart - overall attrition
labels = ['Stayed', 'Left']
sizes = [1233, 237]
colors = ['#66b3ff', '#ff9999']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', startangle=90)
plt.title('Employee Attrition Overview')
plt.show()
```

```python
# Bar chart - attrition by department
dept_counts = df.groupby('Department')['AttritionFlag'].sum()

plt.figure(figsize=(8, 5))
plt.bar(dept_counts.index, dept_counts.values, color=['#66b3ff', '#ff9999', '#99ff99'])
plt.title('Number of Employees Who Left by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=15)
plt.show()
```

```python
# Line chart - attrition rate by age group
age_data = df.groupby('AgeGroup', observed=False)['AttritionFlag'].mean() * 100

plt.figure(figsize=(8, 5))
plt.plot(age_data.index, age_data.values, 
         color='#ff9999', marker='o', linewidth=2)
plt.title('Attrition Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Attrition Rate (%)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Horizontal bar chart - attrition by job role
jobrole_data = df.groupby('JobRole')['AttritionFlag'].mean() * 100
jobrole_data = jobrole_data.sort_values(ascending=True)

plt.figure(figsize=(9, 6))
plt.barh(jobrole_data.index, jobrole_data.values, color='#ff9999')
plt.title('Attrition Rate by Job Role')
plt.xlabel('Attrition Rate (%)')
plt.ylabel('Job Role')
plt.tight_layout()
plt.show()
```

```python
# Bar chart - average salary by attrition
salary_data = df.groupby('Attrition')['MonthlyIncome'].mean()

plt.figure(figsize=(7, 5))
plt.bar(salary_data.index, salary_data.values,
        color=['#66b3ff', '#ff9999'], width=0.5)
plt.title('Average Monthly Salary by Attrition')
plt.xlabel('Attrition')
plt.ylabel('Average Monthly Income ($)')
plt.xticks([0, 1], ['Stayed', 'Left'])

# Add value labels on top of bars
for i, value in enumerate(salary_data.values):
    plt.text(i, value + 100, f'${value:.0f}', 
             ha='center', fontsize=11)

plt.tight_layout()
plt.show()
```

```python
# Function to analyze any column automatically
def analyze_attrition(column_name):
    print(f"=== Attrition Analysis by {column_name} ===")
    result = df.groupby(column_name)['AttritionFlag'].mean() * 100
    result = result.sort_values(ascending=False).round(2)
    print(result)
    print()

# Now call it for multiple columns automatically!
analyze_attrition('Department')
analyze_attrition('JobRole')
analyze_attrition('OverTime')
analyze_attrition('MaritalStatus')
analyze_attrition('BusinessTravel')
```

```python
# Automatically generate and save a report
def generate_report():
    print("Generating report...")
    
    # Create summary data
    summary = {
        'Total Employees'        : len(df),
        'Employees Who Left'     : df['AttritionFlag'].sum(),
        'Employees Who Stayed'   : (df['AttritionFlag'] == 0).sum(),
        'Attrition Rate %'       : round(df['AttritionFlag'].mean() * 100, 2),
        'Average Age'            : round(df['Age'].mean(), 1),
        'Average Salary'         : round(df['MonthlyIncome'].mean(), 2),
        'Lowest Salary'          : df['MonthlyIncome'].min(),
        'Highest Salary'         : df['MonthlyIncome'].max(),
    }
    
    # Print report
    print("\n=============================")
    print("   IBM HR ATTRITION REPORT   ")
    print("=============================")
    for key, value in summary.items():
        print(f"{key:25} : {value}")
    print("=============================")
    
    # Save to CSV
    report_df = pd.DataFrame(list(summary.items()), 
                             columns=['Metric', 'Value'])
    report_df.to_csv('C:/Users/LENOVO/Downloads/IBM_HR_Report.csv', 
                     index=False)
    print("\n✅ Report saved to Downloads folder!")

# Run the function
generate_report()
```
